# Ejemplo: abrir aplicación y capturar screenshot

```python
import subprocess
from autotest.window import wait_for_window

# Abrir gedit
subprocess.Popen(["gedit"])

# Esperar a que la ventana esté visible
geom = wait_for_window("gedit", timeout=10)
print("Ventana encontrada:", geom)
```

```python
from autotest.screenshot import capture_window

capture_window("gedit_capture.png", window_title="gedit")
```
